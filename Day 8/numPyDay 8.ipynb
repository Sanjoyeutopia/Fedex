```


# Assuming 365 days for simplicity, representing one year
num_days = 366


dates = np.arange(np.datetime64('2020-01-01'), np.datetime64('2021-01-01'))
temperatures = np.random.normal(20, 5, num_days)  # Avg temp of 20Â°C with some variation
precipitations = np.random.uniform(0, 20, num_days)  # Random precipitation between 0 and 20mm

# Combine into a single structured array
dtype = [('date', 'datetime64[D]'), ('temperature', float), ('precipitation', float)]
weather_data = np.zeros(num_days, dtype=dtype)
print(weather_data)

weather_data['date'] = dates
weather_data['temperature'] = temperatures
weather_data['precipitation'] = precipitations
#print(weather_data)
#print(precipitations)
```

```
#Calculate monthly average temperatures:
##get unique months
        ###create an numpy array for months
        ###run the unique operations
months = np.array([date.astype('datetime64[M]') for date in weather_data['date']])
unique_months = np.unique(months)
print(unique_months)
###create the average for those months
###mean date.astype("")
for month in unique_months:
    avg_temp = np.mean(weather_data['temperature'])
    
    print(month, avg_temp)



```

```
##createing grades for the learners 
##mean median and stndard deviation from the numpy functions
###the learners who need most of the help
###where clause in np
grades = np.random.randint(50, 100, size=10)
mean=np.mean(grades)
median=np.median(grades)
std=np.std(grades)
print(grades)
print(mean, median, std)

####learners need most help
thresholdvalue=mean-std
helpStudents=np.where(grades<thresholdvalue)[0]
print(helpStudents)

####overachievers : identify the 10% of the class
##overachievers:  identify the top 10
percentile=np.percentile(grades,10)
print(percentile)
top_10_percent_threshold = np.percentile(grades, 90)
overachiever=np.where(grades>=top_10_percent_threshold)[0]
print(overachiever)
###numpy implementation
###numpy functions
###2D array implementations


```
