```

user_activities = [
    {"user_id": "001", "login_count": 10, "pages_visited": 25, "time_spent_minutes": 120, "last_login": "2024-01-15 08:30:00"},
    {"user_id": "002", "login_count": 8,  "pages_visited": 18, "time_spent_minutes": 90,  "last_login": "2024-01-14 09:20:00"},
    {"user_id": "003", "login_count": 15, "pages_visited": 30, "time_spent_minutes": 150, "last_login": "2024-01-16 10:45:00"},
    {"user_id": "004", "login_count": 5,  "pages_visited": 10, "time_spent_minutes": 50,  "last_login": "2024-01-13 11:00:00"},
    {"user_id": "005", "login_count": 12, "pages_visited": 24, "time_spent_minutes": 180, "last_login": "2024-01-16 08:00:00"},
    {"user_id": "006", "login_count": 20, "pages_visited": 45, "time_spent_minutes": 220, "last_login": "2024-01-17 12:30:00"},
    {"user_id": "007", "login_count": 3,  "pages_visited": 5,  "time_spent_minutes": 30,  "last_login": "2024-01-12 07:15:00"},
    {"user_id": "008", "login_count": 7,  "pages_visited": 15, "time_spent_minutes": 70,  "last_login": "2024-01-15 06:45:00"},
    {"user_id": "009", "login_count": 9,  "pages_visited": 20, "time_spent_minutes": 100, "last_login": "2024-01-14 08:30:00"},
    {"user_id": "010", "login_count": 11, "pages_visited": 22, "time_spent_minutes": 110, "last_login": "2024-01-16 07:50:00"}
]
### 1. Total Pages Visited by All Users

def total_pages_visited(data):
    ####Calculate the total number of pages visited by all users.####
    return sum(map(lambda user: user['pages_visited'], data))
####

### 2. Average Pages Visited Per Login

def average_pages_per_login(data):
    ####Calculate the average number of pages visited per login.####
    total_pages = sum(map(lambda user: user['pages_visited'], data))
    total_logins = sum(map(lambda user: user['login_count'], data))
    return total_pages / total_logins if total_logins > 0 else 0
####


### 3. Count Users with Single Login

def count_single_login_users(data):
    ####Count the number of users who have logged in only once.####
    return len(list(filter(lambda user: user['login_count'] == 1, data)))
####

### 4. Average Login Count

def average_login_count(data):
    ####Calculate the average login count across all users.####
    return sum(map(lambda user: user['login_count'], data)) / len(data) if data else 0
####

### 5. Find Latest Login Date

def find_latest_login(data):
    ####Find the most recent login date among all users.####
    return max(data, key=lambda user: user['last_login'])['last_login']
####

### 6. Time Series Analysis of Logins

def login_time_series(data):
    ####Count the number of logins per day.####
    from collections import defaultdict
    login_counts = defaultdict(int)
    for login_date in map(lambda user: user['last_login'].split(' ')[0], data):
        login_counts[login_date] += 1
    return dict(login_counts)
####

### 7. Segment Users Based on Time Spent

def segment_users_by_time(data):
    ####Categorize users into High, Medium, and Low based on time spent.####
    categories = {'High': [], 'Medium': [], 'Low': []}
    for user in data:
        category = ('High' if user['time_spent_minutes'] > 120 
                    else 'Medium' if user['time_spent_minutes'] > 60 
                    else 'Low')
        categories[category].append(user['user_id'])
    return categories
####

### 8. Users with Increasing Activity

def users_increasing_activity(data):
    ####Identify users with an increasing activity trend.####
    # Will look for ML data. This is taking an examle of login count being 10
    return [user['user_id'] for user in data if user['login_count'] > 10]
####

### 9. Find Least Engaged Users

def least_engaged_users(data):
    ####Identify users with the lowest page visits.####
    min_pages = min(map(lambda user: user['pages_visited'], data))
    return [user['user_id'] for user in data if user['pages_visited'] == min_pages]
####

### 10. Most Active Users by Login Count

def most_active_users(data, threshold):
    ####Filter users with a login count above a specified threshold.####
    return [user['user_id'] for user in data if user['login_count'] > threshold]
####

### 11. User Activity on Latest Login

def latest_login_activity(data):
    ####Extract and transform data of users' latest login activity.####
    return [{'user_id': user['user_id'], 'last_login_activity': user['last_login']} for user in data]
####

### 12. Calculate Total Time Spent

def total_time_spent(data):
    ####Aggregate the total time spent by all users.####
    return sum(map(lambda user: user['time_spent_minutes'], data))
####

### 13. Aggregated User Metrics

def aggregate_user_metrics(data):
    ####Compute aggregated metrics like total login count and total pages visited.####
    total_logins = sum(map(lambda user: user['login_count'], data))
    total_pages = sum(map(lambda user: user['pages_visited'], data))
    return {"total_logins": total_logins, "total_pages": total_pages}
####

### 14. Average Pages Visited Per Login

def average_pages_per_login(data):
    ####Calculate the average pages visited per login.####
    total_pages = sum(map(lambda user: user['pages_visited'], data))
    total_logins = sum(map(lambda user: user['login_count'], data))
    return total_pages / total_logins if total_logins > 0 else 0
####

### 15. Anomaly Detection in Page Visits



def detect_anomaly_users(data, threshold):
    ####Identify users whose pages visited count is significantly different from the average.####
    avg_pages_visited = sum(map(lambda user: user['pages_visited'], data)) / len(data)
    return [user['user_id'] for user in data if abs(user['pages_visited'] - avg_pages_visited) > threshold]
####



# 1. Total Pages Visited by All Users
print("Total Pages Visited by All Users:", total_pages_visited(user_activities))
print("##############################")

# 2. Average Pages Visited Per Login
print("Average Pages Visited Per Login:", average_pages_per_login(user_activities))
print("##############################")

# 3. Count Users with Single Login
print("Number of Users with Single Login:", count_single_login_users(user_activities))
print("##############################")

# 4. Average Login Count
print("Average Login Count:", average_login_count(user_activities))
print("##############################")

# 5. Find Latest Login Date
print("Latest Login Date:", find_latest_login(user_activities))
print("##############################")

# 6. Time Series Analysis of Logins
print("Login Time Series:", login_time_series(user_activities))
print("##############################")

# 7. Segment Users Based on Time Spent
print("User Segmentation by Time Spent:", segment_users_by_time(user_activities))
print("##############################")

# 8. Users with Increasing Activity
print("Users with Increasing Activity:", users_increasing_activity(user_activities))
print("##############################")

# 9. Find Least Engaged Users
print("Least Engaged Users:", least_engaged_users(user_activities))
print("##############################")

# 10. Most Active Users by Login Count
threshold = 10  # Define a threshold for most active users
print("Most Active Users (Threshold: {}):".format(threshold), most_active_users(user_activities, threshold))
print("##############################")

# 11. User Activity on Latest Login
print("Latest Login Activity of Users:", latest_login_activity(user_activities))
print("##############################")

# 12. Calculate Total Time Spent
print("Total Time Spent by All Users:", total_time_spent(user_activities))
print("##############################")

# 13. Aggregated User Metrics
print("Aggregated User Metrics:", aggregate_user_metrics(user_activities))
print("##############################")
# 14. Average Pages Visited Per Login (Same as function 2)
print("Average Pages Visited Per Login:", average_pages_per_login(user_activities))
print("##############################")

# 15. Anomaly Detection in Page Visits
anomaly_threshold = 10  # Define an anomaly threshold
print("Anomaly Detection in Page Visits (Threshold: {}):".format(anomaly_threshold), detect_anomaly_users(user_activities, anomaly_threshold))
print("####################")

def users_over_100_minutes(data):
    return list(filter(lambda user: user['time_spent_minutes'] > 100, data))

# Assuming 'user_activities' contains your user data
over_100_minute_users = users_over_100_minutes(user_activities)
print("Users Who Spent Over 100 Minutes:", [user['user_id'] for user in over_100_minute_users])




```
